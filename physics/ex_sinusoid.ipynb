```python
import numpy as np
import matplotlib as mb
import matplotlib.pyplot as plt

def get_sinusoid (t, **kwargs) :
    if "freq" in kwargs:
        radial_speed = kwargs["freq"]
    else :
        radial_speed = 1
    
    if "phase" in kwargs:
        phase = kwargs["phase"]
    else :
        phase = 0
        
    if "amplitude" in kwargs:
        amplitude = kwargs["amplitude"]
    else :
        amplitude = 1
        
    if "period" in kwargs:
        radial_speed = 2*np.pi/kwargs["period"]
        
    return  amplitude*np.sin(radial_speed*t+phase)
        
```

```python
t = np.arange(0,10,0.001)

f = 4
T = 0.5*np.pi

y0 = get_sinusoid(t, freq=f)
y1 = get_sinusoid(t, period=T)

plt.plot(t,y0,'r',label='red : {}radian/s'.format(f))
plt.plot(t,y1,'b',label='blue : {} s'.format(T))
plt.grid()
plt.legend()
print('test {}'.format(2))
```

```python
# interference 

y0 = get_sinusoid(t, freq=3 )
y1 = get_sinusoid(t, freq=3, phase=np.pi/2)
y2=y0+y1
plt.grid()
plt.plot(t,y0,t,y1,t,y2)
```

```python
# cancle out by the out-of-phase

y0 = get_sinusoid(t, freq=3 )
y1 = get_sinusoid(t, freq=3, phase=np.pi)
y2=y0+y1
plt.grid()
plt.plot(t,y0,t,y1,t,y2)
```
