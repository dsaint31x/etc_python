```python
import numpy as np
import matplotlib as mb
import matplotlib.pyplot as plt

k = [0,-1,0,0,0,1,0,0,0,-1,0]
x = [1,3/4,1/2,1/4,0,1/4,1/2,3/4,1]
y = np.convolve(k,x)

fig = plt.figure()
x_axis = np.arange(-2.25,2.5,1/4)
plt.plot(x_axis,y,'o')
```
